```python
#많이 쓰이는 함수(ai)
# append(x): 맨 뒤에 1개 추가
# extend(iterable): 여러 개(반복 가능한 것) 이어붙이기
# insert(i, x): i 위치에 삽입
# pop([i]): i의 원소를 꺼내서 반환 + 리스트에서 삭제 (기본: 마지막)
# remove(x): 값 x 첫 번째 1개 삭제 (없으면 ValueError)
# clear(): 전부 삭제(빈 리스트)
# index(x[, start[, end]]): 값 x의 첫 위치(인덱스) 반환 (없으면 ValueError)
# count(x): 값 x 개수
# sort(key=None, reverse=False): 리스트 자체를 정렬(원본 변경)
# reverse(): 리스트 자체를 뒤집기(원본 변경)
#
# copy(): 얕은 복사(shallow copy)
# len(a): 길이
# sum(a): 합(숫자 리스트)
# min(a), max(a): 최소/최대
# sorted(a, key=..., reverse=...): 정렬된 새 리스트 반환(원본 유지)
# reversed(a): 뒤집힌 “반복자” 반환 (list(reversed(a))로 리스트화)
# enumerate(a, start=0): (인덱스, 값) 형태로 순회
# zip(a, b, ...): 여러 리스트를 묶어서 순회
# list(iterable): iterable을 리스트로 변환
# any(a), all(a): 조건 판정(요소들이 True/False로 평가됨)
```
