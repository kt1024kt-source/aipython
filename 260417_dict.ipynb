```python
# 1) 종료 시 '종료' 입력
while True:
    menu = input("메뉴 입력 (종료 입력 시 끝): ").strip()

    if menu == "종료":
        print("주문이 종료되었습니다.")
        break

    if menu not in menu_prices:
        print("메뉴가 존재하지 않습니다. 다시 입력하세요.")
        continue

    # 2) 메뉴 입력 후 수량 입력(정수, 1 이상)
    qty_raw = input("수량 입력: ").strip()

    if not qty_raw.isdigit() or int(qty_raw) <= 0:
        print("수량은 1 이상의 정수로 입력하세요.")
        continue

    qty = int(qty_raw)

    # 여기부터가 중요: 반드시 while 안에 있어야 함
    if menu in order:
        before = order[menu]
        order[menu] += qty
        print(f"[수량추가] {menu} 기존 {before}개 -> 현재 {order[menu]}개")
    else:
        order[menu] = qty
        print(f"[신규추가] {menu} x {qty}")
```
