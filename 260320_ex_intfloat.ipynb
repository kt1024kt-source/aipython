```python
# 다음 시간에 할 것. 선택구조 만들기 abc 중에 하나 고르기
# if문 (선택, 조건문 )활용 - 입력받은 수가 짝수인지 홀 수인지 판별
# 정수를 입력받아 짝수/홀수를 판별하는 프로그램

n = int(input("정수를 입력하세요: "))

# % 2의 결과가 0이면 짝수, 1이면 홀수
if n % 2 == 0:
    print(f"{n}은(는) 짝수입니다.")
else:
    print(f"{n}은(는) 홀수입니다.")
```

```python
#위 코드 활용 - 정수, 음수 구별 프로그램.
n = int(input("정수를 입력하세요: "))

# % 2의 결과가 0이면 짝수, 1이면 홀수
if n >= 0:
    print(f"{n}은(는) 양수입니다.")
else:
    print(f"{n}은(는) 음수입니다.")
```

```python
#음료 가격
#커피 2000
#라떼 3000
#녹차 2500
#주스 2800
#실행 시 모든 메뉴의 주문 수를 순서대로 각각 물어봄 (단, 이때 0개 주문은 없음)
#커피 1개, 라떼 1개 녹차 1개 주스 1개.
#이후 주문금액 출력
```

```python
#1치 풀이
c = int(input("커피를 몇 개 주문할까요?: "))
cc =c* 2000
l = int(input("라떼를 몇 개 주문할까요?: "))
lc =l* 3000
t = int(input("녹차를 몇 개 주문할까요?: "))
tc =t* 2500
j = int(input("주스를 몇 개 주문할까요?: "))
jc =j* 2800
total = cc + lc + tc + jc
print(f'커피 주문 수 : {c}개')
print(f'라떼 주문 수 : {l}개')
print(f'녹차 주문 수 : {t}개')
print(f'주스 주문 수 : {j}개')
print(f'총 주문 금액{total}원 입니다.')
```

```python
#2차 풀이11
c = int(input("커피를 몇 개 주문할까요?: "))*2000
l = int(input("라떼를 몇 개 주문할까요?: "))*3000
t = int(input("녹차를 몇 개 주문할까요?: "))*2500
j = int(input("주스를 몇 개 주문할까요?: "))*2800
total = c + l + t + j
c /= 2000
l /= 3000
t /= 2500
j /= 2800
print(f'커피 주문 수 : {c}개')
print(f'라떼 주문 수 : {l}개')
print(f'녹차 주문 수 : {t}개')
print(f'주스 주문 수 : {j}개')
print(f'총 주문 금액{total}원 입니다.')

#현재 오버플로우 에러 발생 중.
```
