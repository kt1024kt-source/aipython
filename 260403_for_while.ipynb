```python
import random
answer = random.randint(1,100)
#answer = 1
user = 0
t = 1
while user != answer :
  user = int(input("정답을 입력하세요."))
  if user > answer:
    print ("더 작은 값입니다!")
  if answer > user :
    print("더 큰 값입니다!")
  t+=1
print(f"정답입니다! 정답까지 시도한 횟수:{t}")

```

```python
num = int(input("약수를 알고싶은 수를 입력하세요."))
t = 0
for i in range(1, num+1) :
  if num%i == 0 :
    print(f"{num}의 약수{i}")
    t+=1
print(f"총 약수의 개수{t}개")
```
