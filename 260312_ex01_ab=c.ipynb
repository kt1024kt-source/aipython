```python
# ---------------------------
# 1) 평균 구하기(3과목)
# ---------------------------

# 3과목 점수를 변수에 저장
score1 = 80
score2 = 95
score3 = 70

# 총점과 평균 계산
total = score1 + score2 + score3
average = total / 3

# 평균 결과 출력
print("=== 평균 구하기 ===")
print("총점:", total)
print("평균:", average)


# ---------------------------
# 2) 사칙연산(+,-,*,/)
# ---------------------------

# 계산할 두 수를 변수에 저장
a = 12
b = 5

# 사칙연산 결과를 각각 변수에 저장
add_result = a + b
sub_result = a - b
mul_result = a * b
div_result = a / b

# 사칙연산 결과 출력
print("\n=== 사칙연산 ===")
print("a + b =", add_result)
print("a - b =", sub_result)
print("a * b =", mul_result)
print("a / b =", div_result)
```

```python
# a, b, c를 사용자에게 입력받기
# input()은 글자로 입력되므로 int()로 숫자로 바꿔서 저장
a = int(input("a 값을 입력하세요: "))
b = int(input("b 값을 입력하세요: "))
c = int(input("c 값을 입력하세요: "))

# ---------------------------
# 1) 사칙연산 (a, b만 사용)
# ---------------------------
add_result = a + b
sub_result = a - b
mul_result = a * b
div_result = a / b

print("\n=== 사칙연산(a, b) ===")
print("a + b =", add_result)
print("a - b =", sub_result)
print("a * b =", mul_result)
print("a / b =", div_result)

# ---------------------------
# 2) 평균 구하기 (a, b, c 사용)
# ---------------------------
total = a + b + c
average = total / 3

print("\n=== 평균(a, b, c) ===")
print("총점:", total)
print("평균:", average)
```
