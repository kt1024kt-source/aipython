```python
#복잡한 문제 풀어보기
# 가위바위보 만들기
import random
# 1 가위 / 2 바위 / 3보
# 사람은 각각 둘
user1 = random.randint(1,3)
user2 = random.randint(1,3)

if user1 == 1 :
  user1_w  = "가위"
elif user1 ==2 :
  user1_w  = "가위"
else :
  user1_w = "보"


if user2 == 1 :
  user2_w  = "가위"
elif user2 == 2 :
  user2_w  = "가위"
else :
  user2_w = "보"

#비길때, 가위가 이길 때, 바위가 이길 떄, 보가 이길 때,
if user1 == user2 :
  print(f"유저 1 :{user1_w}")
  print(f"유저 2 :{user2_w}")
  print("비겼습니다!")
elif (user1 == 1) and (user2 == 3) : #user1이 가위일 때.
  print(f"유저 1 :{user1_w}")
  print(f"유저 2 :{user2_w}")
  print("user1 승리!")
elif (user1 == 1) and (user2 == 2) :
  print(f"유저 1 :{user1_w}")
  print(f"유저 2 :{user2_w}")
  print("user2 승리!")
elif (user1 == 2) and (user2 == 1) : #user1이 바위일 때.
  print(f"유저 1 :{user1_w}")
  print(f"유저 2 :{user2_w}")
  print("user1 승리!")
elif (user1 == 2) and (user2 == 3) :
  print(f"유저 1 :{user1_w}")
  print(f"유저 2 :{user2_w}")
  print("user2 승리!")
elif (user1 == 3) and (user2 == 2) : #user1이 보 일 때.
  print(f"유저 1 :{user1_w}")
  print(f"유저 2 :{user2_w}")
  print("user1 승리!")
elif (user1 == 2) and (user2 == 1) :
  print(f"유저 1 :{user1_w}")
  print(f"유저 2 :{user2_w}")
  print("user2 승리!")




```

```python
#복잡한 문제 풀어보기
# 가위바위보 만들기
# 조건식 줄여보기
import random
# 1 가위 / 2 바위 / 3보
# 사람은 각각 둘
user1 = random.randint(1,3)
user2 = random.randint(1,3)

if user1 == 1 :
  user1_w  = "가위"
elif user1 ==2 :
  user1_w  = "바위"
else :
  user1_w = "보"


if user2 == 1 :
  user2_w  = "가위"
elif user2 == 2 :
  user2_w  = "가위"
else :
  user2_w = "보"


if user1 == user2 :
  print(f"유저 1 :{user1_w}")
  print(f"유저 2 :{user2_w}")
  print("비겼습니다!")
#사람 1이 이길 때
elif ((user1 == 1) and (user2 == 3)) or ((user1 == 2) and (user2 == 1)) or ((user1 == 3) and (user2 == 2)) :
  print(f"유저 1 :{user1_w}")
  print(f"유저 2 :{user2_w}")
  print("user1 승리!")
else : #사람 2 가 승리할 경우 :
  print(f"유저 1 :{user1_w}")
  print(f"유저 2 :{user2_w}")
  print("user2 승리!")
```
